```python
import scanpy as sc
import pandas as pd
import numpy as np
```

```python
repeat_times = 0
simple_path = '/home/cavin/jt/benchmark/data/hbc/s1_filtered_cells.h5ad'
celltype_col = 'cell_type'
cell_emb_col = 'X_pca'
pca_components = 50
random_seed=2026 + repeat_times * 200
```
