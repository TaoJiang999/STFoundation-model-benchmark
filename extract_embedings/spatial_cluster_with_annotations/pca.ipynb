```python
import scanpy as sc
import pandas as pd
import numpy as np
```

```python
simple_path = '/home/cavin/jt/benchmark/data/human_lung_cancer/SMI_Lung.h5ad'
celltype_col = 'cell_type'
cell_emb_col = 'X_pca'
pca_components = 50
random_seed=2026
```

```python
adata = sc.read_h5ad(simple_path)
adata
```

```python
sc.pp.pca(adata, n_comps=pca_components,random_state=random_seed)
adata
```

```python
emb = adata.obsm['X_pca']
emb.shape
```

```python
emb_df = pd.DataFrame(
    emb, 
    index=adata.obs_names,
    columns=[f"PCA_dim_{i}" for i in range(emb.shape[1])]
)
save_path = "/home/cavin/jt/benchmark/experiments/embedings/spatial_cluster_with_annotations/PCA_human_lung_cancer_emb.parquet"
emb_df.to_parquet(save_path, compression="zstd")
```
